```python
from activityLabeling_LDS import activityClassifier_LDS, activityClassifier_LDS_db2
from ssm import LDS
import numpy as np
import matplotlib.pyplot as plt
```

# Overview
activityLabeling_* is a set of classes used to featurize time series data with wavelet transformations and infer latent state labels. The class is modeled after sklearn API and intended to provide one line fitting and latent state prediction after construction. The class is also pickleable to allow easy transfer of the prediction set between data sets.

To begin, initialize the the classifier object, here we will demonstrate a classifier using haar wavelet transformation and LDS latent state labeling



```python
nx=10 #number of input features to retain for convolution
square=[1,2,5,9] #input features to take absolute value of before wavelet convolution.
                 #Use for features that are functionally symmetic across positive and negative
a=np.array([5,15,30,]) #wavelet dilations to use in featurization
D=1 #dimensionality of LDS embedding
positive=True #whether to take absolute value of wavelet convolutions for latent space calculation
            #use if interested in the magnitude of change happening

Activity = activityClassifier_LDS(square=square,a=a,nx=nx,D=D,
                                  positive=positive)
```

Load data. Should be in list form with each time series as an element of list

```python
data_file=['X1.npy','X2.npy']
data= [np.load(file) for file in data_file]
```

## Fitting
Easy as can be. Just call fit_raw() function of the activityLabeling class and pass the data set. It will apply your wavelet convolutions and run the fit function for the latent type.

```python
Activity.fit_raw(data,num_iters=10)
```

## Prediction

Predition of latent states is equally convenient. Call predict_raw(data) with the data you want to transfer. Also, if you want the wavelet features for a set of data, this can be returned with waveletTransform(data) 

```python
z,y = Activity.predict_raw(data[0])
wav = Activity.waveletTransform(data[0])
```

```python
en=10000
fig, ax = plt.subplots(nrows=3, figsize=(10,8), sharex=True)

for i in range(2):
    ax[0].plot(data[0][:en,i], lw=.5,alpha=.6)
ax[1].plot(wav[:en,0],lw=.5,alpha=.6)
ax[1].plot(wav[:en,4],lw=.5,alpha=.6)
ax[2].plot(-z[:en])

ax[0].set_ylabel('input features')
ax[1].set_ylabel('Wavelet transform')
ax[2].set_ylabel('Activity (Z)')
ax[2].set_xlabel('time')
fig.savefig('example.png')
```
